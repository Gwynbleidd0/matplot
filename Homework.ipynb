```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
%matplotlib inline
plt.style.use('classic')
```

```python
# 
def func(x,alpha,beta):
    try:
        return((x**beta+alpha**beta)/(x**beta))
    except:
        return 0
fig = plt.figure(figsize = (20,20))#  
plt.subplot(1, 2, 1)#  
x =np.linspace(-10,10,100)#  
plt.plot(x,func(x,1,1),'--r',label='alpha=1,beta=1')
plt.plot(x,func(x,2,1),'--y',label='alpha=2,beta=1')
plt.plot(x,func(x,1,2),'--b',label='alpha=1,beta=2')
plt.plot(x,func(x,2,2),'--g',label='alpha=2,beta=2')
plt.title(" 1")
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
plt.xlabel("x")        
plt.ylabel("y")
plt.subplot(1, 2, 2)
plt.plot(x,func(x,1,0.5),'--r',label='alpha=1,beta=0.5')
plt.plot(x,func(x,1,-0),'--y',label='alpha=1,beta=-0')
plt.plot(x,func(x,1,-1.5),'--b',label='alpha=1,beta=-1.5')
plt.title(" 2")
plt.xlabel("x")        
plt.ylabel("y")
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig.savefig('pic.png')

```

```python
#1
iris = load_iris()
features = iris.data.T
rng = np.random.RandomState(0)
plt.scatter(features[3], features[2], alpha=0.5,
s=100*rng.rand(100), c=iris.target, cmap='viridis')
plt.ylabel(iris.feature_names[2])
plt.xlabel(iris.feature_names[3])
plt.colorbar()
```

```python
#2
x = np.linspace(0, 10, 10)
y = np.sin(x)
plt.plot(x, y,'d', color = (0.5,0.1,0.5))
plt.plot(x, y, ':r',label='line');
rng = np.random.RandomState(1)
for marker in ['d', 'x', '+']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
    label="marker='{0}'".format(marker))
    plt.legend(numpoints=1)
    plt.ylim(0,1)
    plt.xlim(0, 1.2)
plt.plot(x, y, ':r',label='line');
```

```python
#3D 
from mpl_toolkits import mplot3d
```

```python
#1
def f(x,y):
    return(np.cos(x**2)+np.sin(y**2))
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlabel('x')
ax.view_init(50, 35) 
```

```python
#2
fig = plt.figure()
ax = plt.axes(projection='3d')
zline = np.linspace(-15, 15, 1000)
yline = np.tan(zline)
xline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'red')
```
